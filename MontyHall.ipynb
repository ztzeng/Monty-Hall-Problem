# Simulation

```python
import random
from matplotlib import pyplot as plt
from IPython.core.pylabtools import figsize
```

```python
def simulation_for_switch(n): # n is the number of games played
    count_win = 0
    for i in range(n):
        doors = [1,2,3] 
        prize = random.choice(doors)
        guest_choice = random.choice(doors)
        if guest_choice != prize: # Monty will not open the prize door, the un_opened door is the prize door, also swithced door 
            count_win +=1
        if guest_choice == prize: # leave the prize door, lose
            count_win +=0
    return count_win*1.0/n
```

```python
win_prc = []
N = []
for i in range(1,1000): ## outer loop from 1 to 1000
    win_prc.append(simulation_for_switch(i))
    N.append(i)
```

```python
figsize(12.5, 3.5)
plt.plot(N,win_prc,alpha=0.85)
plt.ylabel("proportion of wining when switch")
plt.xlabel("number of games played")
plt.title("simulation for switch")
plt.ylim([0,1])
plt.xlim([0,1000])
```
